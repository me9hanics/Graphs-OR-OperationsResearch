# Minimum cost circulation problem

An implementation based on Goldberg-Tarjan 1989 algorithm, described in the 2004 book of Schrijver (Combinatorial Optimization) of the minimum cost circulation problem in the *computation_functions.py* file.

```python
import numpy as np
import networkx as nx
import computation_functions

#Nodes: A, B, C, D, 
G = nx.DiGraph()
G.add_nodes_from(["A", "B", "C", "D"])
G = nx.DiGraph()
G.add_nodes_from(["A", "B", "C", "D"])
G.add_edges_from([
    ("A", "B", {'capacity': 3, 'flow': 0, 'cost': -1}),
    ("A", "C", {'capacity': 1, 'flow': 0, 'cost': -1}),
    ("A", "D", {'capacity': 4, 'flow': 0, 'cost': -1}),
    ("B", "D", {'capacity': 3, 'flow': 0, 'cost': -1}),
    ("C", "D", {'capacity': 1, 'flow': 0, 'cost': -1}),
    ("D", "A", {'capacity': 7, 'flow': 0, 'cost': -1})
])
```

Draw the graph:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='lightblue')
nx.draw_networkx_labels(G, pos)
for u, v, d in G.edges(data=True):
    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], width=d['capacity'],
                           connectionstyle='arc3,rad=0.3')
labels = nx.get_edge_attributes(G, 'capacity')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, label_pos=0.7)

plt.show()
```
